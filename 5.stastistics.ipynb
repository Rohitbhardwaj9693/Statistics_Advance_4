```python
# Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
# given a sample from each population.



import numpy as np
import scipy.stats as stats

sample1 = np.array([68, 72, 65, 71, 73, 67, 70, 74, 72, 69])
sample2 = np.array([63, 66, 62, 68, 65, 67, 61, 70, 68, 64])



alpha = 0.05
t_statisttic,p_value = stats.ttest_ind(sample1,sample2)


print('sample 1 mean',np.mean(sample1))
print('sample2 mean',np.mean(sample2))
# print('t-stastistics',t_statisttic)
print("p-value",p_value)

if p_value< alpha:
    print('Reject the null hypothesis')
else:
     print("fail to reject the null hypothesis")
```

```python
# Q6: What is a null and alternative hypothesis? Give some examples.


# Null Hypothesis (H0):

# The null hypothesis is a statement that represents a default or conservative
# assumption, suggesting that there is no effect, no difference, or no 
# relationship in the population.

# It often contains an equality sign (e.g., =, ≤, ≥) and reflects the status
# quo or a lack of change.
# The null hypothesis is what you aim to test against to determine if there
# is enough evidence to reject it in favor of the alternative hypothesis.

# Examples:
# For a coin toss experiment: H0: The coin is fair (the probability of hea P(heads), is 0.5)


# Alternative Hypothesis (Ha or H1):

# The alternative hypothesis is a statement that contradicts the null hypothesis
# and represents the claim or effect you want to investigate or demonstrate.
# It often contains inequality signs (e.g., ≠, <, >) and reflects a change,
# effect, or difference in the population.
# The alternative hypothesis is what you aim to provide evidence for when
# conducting a hypothesis test.

# Examples:
# For a coin toss experiment: Ha: The coin is biased (P(heads) ≠ 0.5)
```
